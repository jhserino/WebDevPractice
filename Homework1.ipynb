```python
#Import the matplotlib.pyplot module so that we can display functions
#in a plot
#Import the numpy module for mathematical functions such as sin(x)
import matplotlib.pyplot as plt
import numpy as np
import math #for pi
```

```python
x=np.linspace(0,2*math.pi,75) # 75 Equally spaced points between  (0,2pi)
y=np.sin((x))*x
plt.figure()
plt.plot(x,y)
plt.show()
```
